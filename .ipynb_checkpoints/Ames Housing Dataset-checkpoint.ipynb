### Model

```python
train.head()
```

```python
X=train.drop(columns="SalePrice")
```

```python
y=train["SalePrice"].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
features=["1stFlrSF"]
```

```python
def log_mean_squared_error(truth,predictions):
    ltruth=np.log(truth)
    lpredictions=np.log(predictions)
    return np.sqrt(sum((ltruth-lpredictions) ** 2)/len(truth))
```

```python
lmse_scorer=make_scorer(log_mean_squared_error,greater_is_better=False)
```

```python
def cross_validation(regr=LinearRegression(),X_train=X_train,y_train=y_train, features=features,scoring=lmse_scorer,
                   table=True):
    
    
    name=regr.__class__.__name__
    outcomes=cross_validate(regr,X_train[features],y_train,scoring=scoring)
    mean_test_score=outcomes["test_score"].mean()
    test_std=outcomes["test_score"].std()
    mean_train_score=outcomes["train_score"].mean()
    train_std=outcomes["train_score"].std()
    
    
    regr_results=pd.DataFrame({"Regressor":name,"Mean train score":mean_train_score,"Train std":train_std,
                              "Mean test score":mean_test_score,"test_std":test_std},index=[1])
    print ("features:" + str(features))
    if table:
        display(regr_results)
    
```
